## **Import Libraries**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as ex
import warnings
warnings.filterwarnings('ignore')
```

```python
data=pd.read_csv("C:/Users/sachi/Downloads/Amazon-product_reviews_analysis.csv")
print(data)
```

```python
df=pd.DataFrame(data)
print(df)
```

```python
# Datasets Infromation

df.shape
df.info()
df.describe(include='all')
```

```python
# Remove Duplicates Rows.

df.drop_duplicates(inplace=True)
print("After removing duplicates:", df.shape)
```

```python
df.isnull().sum()
```

```python
# Drop rows with missing review text
df = df.dropna(subset=['review_text'])

# Fill missing helpful_votes with 0
df['helpful_votes'] = df['helpful_votes'].fillna(0)

# Fill missing category with "Unknown"
df['category'] = df['category'].fillna("Unknown")
```

```python
# Convert Date column into proper format.

df['review_date'] = pd.to_datetime(df['review_date'], errors='coerce')
df['year'] = df['review_date'].dt.year
df['month'] = df['review_date'].dt.month
```

```python
# Clear Column Raring.

df['rating'] = pd.to_numeric(df['rating'], errors='coerce')
df = df[df['rating'].between(1, 5)]  # Keep only valid ratings
```

```python
# Outlier Detections.

Q1 = df['helpful_votes'].quantile(0.25)
Q3 = df['helpful_votes'].quantile(0.75)
IQR = Q3 - Q1

lower = Q1 - 1.5*IQR
upper = Q3 + 1.5*IQR

df = df[(df['helpful_votes'] >= lower) & (df['helpful_votes'] <= upper)]
```

## **Q.1 Average Rating Distributions.**

```python
plt.figure(figsize=(8,5))
sns.countplot(x='rating', data=df, palette="viridis")
plt.title("Rating Distribution")
plt.xlabel("Rating")
plt.ylabel("Count")
plt.show()
```

## **Q.2 Average Rating by category.**

```python
plt.figure(figsize=(12,5))
category_rating = df.groupby('category')['rating'].mean().sort_values()

sns.barplot(x=category_rating.index, y=category_rating.values)
plt.xticks(rotation=90)
plt.title("Average Rating by Category")
plt.ylabel("Average Rating")
plt.show()
```

## **Q.3 Number of reviews per year.**

```python
plt.figure(figsize=(10,5))
sns.countplot(x='year', data=df)
plt.title("Review Count by Year")
plt.show()
```

## **Q.4 Distributions of helpful votes.** 

```python
plt.figure(figsize=(8,5))
sns.histplot(df['helpful_votes'], bins=30)
plt.title("Distribution of Helpful Votes")
plt.show()
```

## **Q.5 How do customer ratings vary across different products or categories.** 

```python
plt.figure(figsize=(12,5))How do customer ratings vary across different products or categories?
sns.boxplot(x='category', y='rating', data=df)
plt.xticks(rotation=90)
plt.title("Rating Variation Across Categories")
plt.show()

df.groupby('product_id')['rating'].mean().sort_values().head(10)
```

## **Q.6 How do customer ratings vary across different products or categories.**

```python
rating_sales = df.groupby('rating')['sales'].mean()

plt.plot(rating_sales.index, rating_sales.values)
plt.title("Impact of Rating on Sales")
plt.xlabel("Rating")
plt.ylabel("Average Sales")
plt.show()
```
